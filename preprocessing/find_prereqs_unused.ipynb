```python
BASIC_PATTERN = r"[A-Z]+\s*[\d]+[A-Z]?"

# current bugs:
## CS 103B/X (catch CS103X)
SLASH_PATTERN = r"([A-Z]+\s+[0-9]+)(?:[A-Z]?/)([A-Z])" # not tested

## LINGUIST 130A or B (catch LINGUIST130B)
OR_PATTERN = r""

## Prerequisites: 103 or 103B/X (no dept name, need to pull dept
# from course.subject if there's a lone number with no preceding
# capital letters)
NODEPT_PATTERN = r""

## exposure to basic concepts in probability (ignore descriptions
# with no reference to specific courses?)

# OR/AND shouldn't matter, just count anything mentioned in the
# description as a possible prereq, fit models with only those as
# predictors and measure accuracy



# pass in a course object (includes year), returns the prereqs
# found by scraping description with regex patterns
def get_prereqs_from_desc(course):
    # cut off text before the first word "Prereq" because sometimes
    # the course itself is mentioned in the description, or there
    # could be random numbers that are interpreted as a course
    desc = course.description
    if desc.find("Prereq") == -1 and desc.find("Recomm") == -1:
        return([])
    
    # try to split on "Prereq" because that usually comes first
    if desc.find("Prereq") != -1:
        desc = desc.split("Prereq", 1)[1]
    # if there is no "Prereq" there must be "Recomm"
    elif desc.find("Recomm") != -1:
        desc = desc.split("Recomm", 1)[1]
    
    prereqs = []
    prereqs += re.findall(BASIC_PATTERN, desc)
    clean_prereqs = ["".join(prereq.split()) for prereq in prereqs]
    return(clean_prereqs)
```

```python
for course in CS221:
    print(course.year, course.subject, course.code, get_prereqs_from_desc(course))
```
