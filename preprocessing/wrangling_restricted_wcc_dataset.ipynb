# Creating restricted_wcc dataset (only courses from WCC, no summer, keep only row from last five years or from a student with a course in the last five years)

```python
WCC_PATH = "../data/weakly_connected_enrollments.fthr"
```

```python
wcc_df = pd.read_feather(WCC_PATH)
wcc_df.columns
```

```python
wcc_df
```

```python
# major data is sparse because it's declared major at the time of taking that class only

wcc_df[["major1", "major2", "major3", "major4", "major5", "major6", "major7"]]
```

```python
# we could potentially run into problems with cross listed courses
print(wcc_df["course_name"].value_counts()["LINGUIST 188"])
print(wcc_df["course_name"].value_counts()["LINGUIST 288"])
print(wcc_df["course_name"].value_counts()["CS 224U"])
```

```python
# drop the weird empty duplicate rows
wcc_df.drop(wcc_df[wcc_df["unt_taken"] == 0].index, inplace=True)
wcc_df
```

```python
# removing some unneeded columns and creating no-space "course_name" column
wcc_df = wcc_df.drop(['ssr_component', 'crse_acad_group', 'crse_grade_off', 'last_upd_dt_stmp',
                                'last_upd_tm_stmp', 'class_nbr',
                                'grading_basis_dt', 'effdt', 'unt_prgrss',
                                'grading_scheme_enr', 'crse_id',
                                'major1', 'major2', 'major3', 'major4', 'major5', 'major6', 'major7', 'oprid', 'course_name'], axis=1)
wcc_df["course_name"] = wcc_df["subject"] + wcc_df["catalog_nbr"]
wcc_df
```

```python
# filter to keep only students with at least one course in strm >= 1162
keep_students = wcc_df[wcc_df["strm"] >= 1162]["emplid"].unique()
print(len(keep_students))
print(keep_students)

wcc_df.drop(wcc_df[(wcc_df["strm"] < 1162) & (~wcc_df["emplid"].isin(keep_students))].index, inplace=True)
wcc_df
```

```python
# drop summer courses
wcc_df.drop(wcc_df[wcc_df["strm"] % 10 == 8].index, inplace=True)
wcc_df
```

```python
# don't include dropped courses
wcc_df.drop(wcc_df[wcc_df["stdnt_enrl_status"] == "D"].index, inplace=True)
wcc_df
```

```python
# don't include courses not taken for a letter or the grade isn't recorded yet
LETTER_GRADES = ["A+", "A", "A-", "B+", "B", "B-", "C+", "C", "C-", "D+", "D", "D-", "NP", "W"]

wcc_df.drop(wcc_df[~wcc_df["crse_grade_input"].isin(LETTER_GRADES)].index, inplace=True)
wcc_df
```

```python
print(len(wcc_df), "unique course records")
print(len(wcc_df["emplid"].unique()), "unique students")
print(len(wcc_df["course_name"].unique()), "unique courses")
```
